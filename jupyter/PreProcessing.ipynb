```python
import pandas as pd
import numpy as np
from openpyxl import load_workbook # load data excel
from sklearn.metrics.pairwise import cosine_similarity # cosine similarity
from nltk.corpus import stopwords # preprocessing
from nltk.stem import PorterStemmer # preprocessing bahasa inggris
import string # ya buat string
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory 
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
stemmer = StemmerFactory().create_stemmer()
remover = StopWordRemoverFactory().create_stop_word_remover()
translator = str.maketrans('', '', string.punctuation)

def stemmerEN(text):
    porter = PorterStemmer()
    stop = set(stopwords.words('indonesian')) #stopwods berguna untuk
    text = text.lower()
    text = [i for i in text.lower().split() if i not in stop]
    text = ' '.join(text)
    preprocessed_text = text.translate(translator)
    text_stem = porter.stem(preprocessed_text)
    return text_stem

def preprocessing(text):
    text = text.lower()
    text_clean = remover.remove(text)
    text_stem = stemmer.stem(text_clean)
    text_stem = stemmerEN(text_stem)
    return text_stem
```

```python
setD = pd.read_excel('data berita 2019 fix.xlsx',sep=',')
Doc = setD['judul berita'].astype('str')

preprocessed_docs = []
for stemming in Doc:
    preprocessed_docs.append(preprocessing(stemming))
# add data to excel  
setD["preprocessed_judul"] = preprocessed_docs
#save to excel
setD.to_excel("data berita 2019 fix.xlsx", engine="xlsxwriter", index=False)
dataset = pd.read_excel("data berita 2019 fix.xlsx")
dataset
```
