## TF_IDF Cosine Similarity

```python
from sklearn.feature_extraction.text import CountVectorizer # tf-idf
from sklearn.feature_extraction.text import TfidfTransformer, TfidfVectorizer # tf-idf
from sklearn.decomposition import TruncatedSVD
from sklearn.pipeline import Pipeline
from nltk.corpus import stopwords # preprocessing\
from math import*
import numpy as np
from numpy import linalg as LA
import math

#class engine berugan untuk menghitung VSM dari proses (TFIDF sampai Cosine)
class Engine:
    
    def __init__(self):
        self.cosine_score = []
        self.train_set = []  # Memasukkan Documents
        self.test_set = []  # Memasukkan Query

    def addDocument(self, word): # fungsi untuk menambahkan dokumen dataset ke dalam list train_set
        self.train_set.append(word)

    def setQuery(self, word):  # fungsi untuk menambahkan data query ke dalam list test_Set
        self.test_set.append(word)

    def process_score(self):
        #stopWords = stopwords.words('english') 
        stopWords = stopwords.words('english') 
        vectorizer = CountVectorizer()
        #dari sini mulai proses TF-IDF
        transformer = TfidfTransformer()

        trainVectorizerArray = vectorizer.fit_transform(self.train_set).toarray() 
        # menghitung Bobot dokumen dataset dan uji dan kemudian disimpan dalam bentuk array 
        testVectorizerArray = vectorizer.transform(self.test_set).toarray()
        #dari sini mulai proses cosine similarity
        cx = lambda a, b: round(np.inner(a, b) / (LA.norm(a) * LA.norm(b)), 3) 
        #fungsi tanpa nama untuk normalisasi data dan definisi rumus Cosine Similarity 
        #         print testVectorizerArray
        output = []
        for i in range(0, len(testVectorizerArray)):
            output.append([])

        for vector in trainVectorizerArray:
            # print vector
            u = 0
            for testV in testVectorizerArray:
                #perhitungan Cosine Similarity dalam bentuk vector dari dataset dengan query
                #yang di masukan yang kemudian mengembalikan nilai cosine ke dalam variable
                #cosine_score dalam bentuk list.
                # print testV
                cosine = cx(vector, testV)
                #                 self.cosine_score.append(cosine)
                #                 bulatin = (round(cosine),2)
                output[u].append((cosine))
                u = u + 1
        return output
        # return testVectorizerArray
```

```python
import pandas as pd
setD = pd.read_excel('data berita 2019 fix.xlsx',sep=',')
Doc = setD['preprocessed_judul'].astype('str')

query = ['sekda jabar eks bos lippo cikarang cegah luar negeri']

engine = Engine() #memanggil class diatas

docs = [str(x) for x in Doc] #melakukan perulangan untuk merubah setiap column menjadi string
documentNames = list()

for i, doc in enumerate(docs): #melakukan perulangan untuk mendapatkan setiap index preprocessing dalam excel
    engine.addDocument(doc) #memasukkan function addDocument didalam class Engine
    documentNames.append("Document_{}".format(i+1)) #memasukkan hasil list function addDocument didalam class Engine ke dalam array
for queries in query: #melakukan perulangn untuk mendapatkan setiap index dari query jika query > 1
    engine.setQuery(queries) #inputandata uji

titles_score = engine.process_score() #memanggil function process score pada class engine dari hasil addDocument diatas
ScoreDf = (pd.DataFrame(titles_score)).T #merubah data hasil keluaran menjadi DAtaFrame
ScoreDf.columns = query #mendefinisikan quqery sebagai data training
ScoreDf["Documents"] = documentNames #membuat Document sesuai urutan teratas hasil TF-IDF dan VSM
```

```python
from sklearn.model_selection import cross_val_score
import time

#svm_ = svm.SVC(kernel='linear')
df_listed = list()
for i in query:
    labels = list() #membuat array untuk melabel i bahwa dia hasi VSM > 0 atau < 0 
    #labelss = list()
    for j in ScoreDf[i]: #melakuakan pengecekan hasil VSM
        if j>0.000: #membuat label ketika VSM > 0 menampilkan 1
            labels.append(1)
            #labelss.append(cross_val_score(svm_, query, Doc, cv=10)
        else: #membuat label ketika VSM < 0 menampilkan 0
            labels.append(0)
            #labelss.append(cross_val_score(svm_, query, Doc, cv=10)
    datadf = pd.DataFrame(ScoreDf[i])
    datadf['Documents'] = ScoreDf['Documents']
    datadf['Labels'] = labels
    datadf['Judul'] = setD['judul berita'].values
    datadf['Media'] = setD['media'].values
    datadf['Kategori'] = setD['kategori'].values
    print(datadf)
    df_listed.append(datadf.sort_values(by=[i], ascending=False))
df_listed[0].head(5) #hanya mengambil 5 rangking teratas
```
